```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Creamos un dataset sintetico de prueba
```

```python
X, y = make_regression(n_samples=500, n_features=1, noise=30)
```

```python
plt.scatter(X,y)
plt.title('Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Creamos un modelo de regresión lineal
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
# Determinamos la predicción del modelo
```

```python
y_pred = model.predict(X)
```

```python
# Representación gráfica del modelo
```

```python
u = np.linspace(2,3)
```

```python
v = model.predict(u[:, np.newaxis])
```

```python
plt.plot(u,v, 'r', label= 'modelo')
plt.scatter(X,y,label='dataset')
plt.title('Modelo de regresión lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# Errores
```

```python
# Error Absoluto Medio
```

```python
metrics.mean_absolute_error(y,y_pred)
```

```python
# Error cuadrático Medio
```

```python
metrics.mean_squared_error(y, y_pred)
```

```python
# Raiz del error cuadrático medio
```

```python
np.sqrt(metrics.mean_squared_error(y, y_pred))
```

```python
# Coeficiente R2
```

```python
r2 = metrics.r2_score(y,y_pred)
print(r2)
```

```python
# Coeficiente de determinacion ajustado
```

```python
r2_ajust = 1-((1-r2)*(len(y)-1))/(len(y)-X.shape[1]-1)
print(r2_ajust)
```
